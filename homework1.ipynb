Bài 8

```python
name=input("Tên chủ hộ: ")
old=float(input("Chỉ số điện tháng trước(kWh): "))
new=float(input("Chỉ số điện tháng này(kWh): "))
used=new-old
if used<0:
    print("Chỉ số không hợp lệ!")
else:
    if 0<used<=50:
        cost=used*1984
    elif 50<used<=100:
        cost=used*2050
    elif 100<used<=200:
        cost=used*2380
    elif 200<used<=300:
        cost=used*2998
    elif 300<used<=400:
        cost=used*3350
    elif used>400:
        cost=used*3460
total=cost*1.08
print("Họ và tên: ", name)
print("Số điện tiêu thụ: ", used, "kWh")
print("Tiền điện phải trả là: ", round(total), "vnđ")
```
